```python
import pandas as pd
import calendar

df = pd.read_excel('Call List-2020-05-19.xlsx',sheet_name = 1)
df.head(4)
```

```python
# main output is call_logs (a python dictionary w/ keys by month)
df = df.drop_duplicates('Phone Number')  # remove duplicate numbers
 
date_df = pd.to_datetime(df['Start Time']) # create date filter 
unique_months = set(date_df.dt.month)

call_logs = dict() #populate dictionary based on unique months and call logs
for b in unique_months: 
    date_filtered_df = df.loc[date_df.dt.month == b]
    call_logs[calendar.month_abbr[b]] = date_filtered_df.groupby('Number Name')['Phone Number'].nunique()
```
